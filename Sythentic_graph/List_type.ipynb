This ipynb file plots some common substructures in networkx that could be used to build lego structure in Generator.py \
For complete list: https://networkx.org/documentation/stable/reference/generated/networkx.generators.classic.barbell_graph.html

```python
import networkx as nx
```

```python
# Balanced tree
G = nx.balanced_tree(r=3, h=2, create_using=None) # r: number of children, h: height of tree
nx.draw(G)
```

```python
# barbell_graph (two complete graph connected by a path)
G = nx.barbell_graph(m1=4, m2=3)
nx.draw(G)
```

```python
# Complete graph
G = nx.complete_graph(9)
nx.draw(G)
```

```python
# cycle graph
G = nx.cycle_graph(n=10)
nx.draw(G)
```

```python
# path graph
G = nx.path_graph(n=8)
nx.draw(G)
```

```python
# star graph
G = nx.star_graph(n=10)
nx.draw(G)
```

```python
# wheel graph
G = nx.wheel_graph(n=6)
nx.draw(G)
```

```python
# diamond graph
G = nx.diamond_graph()
nx.draw(G)
```

```python
# house graph
G = nx.house_graph()
nx.draw(G)
```
