### Star graph

```python
G = nx.star_graph(9)
L, D, Gamma = graph_to_mat(G)
group_label = [0] * G.number_of_nodes()
U, latent_features, node_features = clean_feature_generator(group_label, dim_features=20, dim_latent_features=8)
```

```python
f = plt.figure(figsize=(15,12))
ax0 = f.add_subplot(4, 4, 1)
nx.draw(G, with_labels=True)
# symmetric-symmetric elastic net
sse = ss_El_Estimator(alpha=1000, beta=1000, gamma=0, Gamma=Gamma, D=D, method='cp', solver='SCS', deg_crct=True)
sse.fit(node_features, maxiter=500)
ax0 = f.add_subplot(4, 4, 2)
sse.heatplot(xticklabels=False, yticklabels=False, cbar=False)
ax0.set_title('ss-elastic net, lap, Wplot')
ax01 = f.add_subplot(4, 4, 6)
sse.scaling(xticklabels=False, yticklabels=False, cbar=False)
ax01.set_title('ss-elastic net, lap, dbplot')
# asmmetric-symmetric elastic net
ase = as_El_Estimator(alpha=1000, beta=1000, gamma=0, Gamma=Gamma, D=D, method='cp', solver='SCS', deg_crct=True)
ase.fit(node_features, maxiter=500)
ax1 = f.add_subplot(4, 4, 3)
ase.heatplot(xticklabels=False, yticklabels=False, cbar=False)
ax1.set_title('as-elastic net, lap, Wplot')
ax11 = f.add_subplot(4, 4, 7)
ase.scaling(xticklabels=False, yticklabels=False, cbar=False)
ax11.set_title('as-elastic net,lap,  dbplot')
# asmmetric-asymmetric elastic net
aae = aa_El_Estimator(alpha=1000, beta=1000, gamma=0, Gamma=Gamma, D=D, method='cp', solver='SCS', deg_crct=True)
aae.fit(node_features, maxiter=500)
ax2 = f.add_subplot(4, 4, 4)
aae.heatplot(xticklabels=False, yticklabels=False, cbar=False)
ax2.set_title('aa-elastic net, lap, Wplot')
ax21 = f.add_subplot(4, 4, 8)
aae.scaling(xticklabels=False, yticklabels=False, cbar=False)
ax21.set_title('aa-elastic net, lap, dbplot')
# symmetric-symmetric elastic net
sse = ss_El_Estimator(alpha=1000, beta=1000, gamma=0, Gamma=Gamma, D=D, method='cp', solver='SCS', deg_crct=False)
sse.fit(node_features, maxiter=500)
ax0 = f.add_subplot(4, 4, 10)
sse.heatplot(xticklabels=False, yticklabels=False, cbar=False)
ax0.set_title('ss-elastic net, org, Wplot')
ax01 = f.add_subplot(4, 4, 14)
sse.scaling(xticklabels=False, yticklabels=False, cbar=False)
ax01.set_title('ss-elastic net, org, dbplot')
# asmmetric-symmetric elastic net
ase = as_El_Estimator(alpha=1000, beta=1000, gamma=0, Gamma=Gamma, D=D, method='cp', solver='SCS', deg_crct=False)
ase.fit(node_features, maxiter=500)
ax1 = f.add_subplot(4, 4, 11)
ase.heatplot(xticklabels=False, yticklabels=False, cbar=False)
ax1.set_title('as-elastic net, org, Wplot')
ax11 = f.add_subplot(4, 4, 15)
ase.scaling(xticklabels=False, yticklabels=False, cbar=False)
ax11.set_title('as-elastic net,org,  dbplot')
# asmmetric-asymmetric elastic net
aae = aa_El_Estimator(alpha=1000, beta=1000, gamma=0, Gamma=Gamma, D=D, method='cp', solver='SCS', deg_crct=False)
aae.fit(node_features, maxiter=500)
ax2 = f.add_subplot(4, 4, 12)
aae.heatplot(xticklabels=False, yticklabels=False, cbar=False)
ax2.set_title('aa-elastic net, org, Wplot')
ax21 = f.add_subplot(4, 4, 16)
aae.scaling(xticklabels=False, yticklabels=False, cbar=False)
ax21.set_title('aa-elastic net, org, dbplot')
```
